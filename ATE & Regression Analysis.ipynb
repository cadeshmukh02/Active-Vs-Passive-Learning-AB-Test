# Power Analysis

```python
# Desirable effect Size = 10

#merged_survey.describe()

# STD:
# RQ1_understand_material_val - 20
# post_class_recall_val - 24
# post_class_class_participation_val - 33

RQ1_d = 10/20
post_class_recall_d = 10/24
post_class_class_participation_d = 10/33

#Power Analysis

RQ1_n = pg.power_ttest(d=RQ1_d, power=0.8, alpha=0.05, alternative='two-sided')
post_class_recall_n = pg.power_ttest(d=post_class_recall_d, power=0.8, alpha=0.05, alternative='two-sided')
post_class_class_participation_n = pg.power_ttest(d=post_class_class_participation_d, power=0.8, alpha=0.05, alternative='two-sided')

print(RQ1_n, post_class_recall_n, post_class_class_participation_n)
```

# TTest, Significance and P-values

```python
import numpy as np
import pandas as pd
import seaborn as sns
import pingouin as pg
pg.options['round.column.CI95%'] = 6
```

```python
merged_survey = pd.read_csv('Merged_survey.csv')
pre_survey = pd.read_csv('pre_survey.csv')
pre_survey = pre_survey[pre_survey['pre_class_incomplete'] == 0]
merged_survey.info()
```

```python
merged_survey.head()
```

```python
pre_treatment = pre_survey[pre_survey['Treatment'] == 1]
pre_control = pre_survey[pre_survey['Treatment'] == 0]

merged_treatment = merged_survey[merged_survey['Treatment'] == 1]
merged_control = merged_survey[merged_survey['Treatment'] == 0]


#RQ1
t_test = pg.ttest(pre_treatment['RQ1_understand_material_val'], pre_control['RQ1_understand_material_val'],
                  paired=False,
                  alternative='two-sided')
print("The p-value result for Understanding the material using ttest:")
print(t_test["p-val"])

#Post_class_recall_val
t_test = pg.ttest(merged_treatment['post_class_recall_val'], merged_control['post_class_recall_val'],paired=False,
                  alternative='two-sided')

print("The p-value result for Recalling the material using ttest:")
print(t_test["p-val"])


#Post_class_class_participation
t_test = pg.ttest(merged_treatment['post_class_class_participation_val'], merged_control['post_class_class_participation_val'],paired=False,
                  alternative='two-sided')

print("The p-value result for Participation in the class using ttest:")
print(t_test["p-val"])
```

```python
import statsmodels.formula.api as smf
from stargazer.stargazer import Stargazer

# Fit the regression model
model_1 = smf.ols("RQ1_understand_material_val ~ Treatment", data=merged_survey).fit(cov_type='HC1')

# Stargazer to display results
Stargazer([model_1])
```
